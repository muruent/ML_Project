```python
#Handling missing (NA) values
```

```python
#looking into original description
#https://ww2.amstat.org/publications/jse/v19n3/Decock/DataDocumentation.txt
```

```python
#NA means "no alley access", replacing with None
combined_data["Alley"] = combined_data["Alley"].fillna("None")
#NA means "no fireplace", replacing with None
combined_data["FireplaceQu"] = combined_data["FireplaceQu"].fillna("None")
#replacing with most common
combined_data['SaleType'] = combined_data['SaleType'].fillna(combined_data['SaleType'].mode()[0])
combined_data["PoolQC"] = combined_data["PoolQC"].fillna("None")
combined_data["MiscFeature"] = combined_data["MiscFeature"].fillna("None")
combined_data["Alley"] = combined_data["Alley"].fillna("None")
combined_data["Fence"] = combined_data["Fence"].fillna("None")
combined_data["FireplaceQu"] = combined_data["FireplaceQu"].fillna("None")
```

```python
#filling front lawn area with median
combined_data["LotFrontage"] = combined_data.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
#replacing with None
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    combined_data[col] = combined_data[col].fillna('None')
    #Replacing missing numerical data with 0 
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    combined_data[col] = combined_data[col].fillna(0)
    # these are categorical features related to  basement, hence filling with None.
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    combined_data[col] = combined_data[col].fillna('None')
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    combined_data[col] = combined_data[col].fillna(0)
```

```python
#NA means no filling with none for type and 0 for area.
combined_data["MasVnrType"] = combined_data["MasVnrType"].fillna("None")
combined_data["MasVnrArea"] = combined_data["MasVnrArea"].fillna(0)
```

```python
#MS Zoning (Nominal): Identifies the general zoning classification of the sale. replacing NA with most common value'RL'
counter=collections.Counter(combined_data['MSZoning'])
print(counter.most_common())
combined_data['MSZoning'] = combined_data['MSZoning'].fillna(combined_data['MSZoning'].mode()[0])
```

```python
combined_data['Electrical'] = combined_data['Electrical'].fillna(combined_data['Electrical'].mode()[0])
#KitchenQual (Ordinal): Kitchen quality - replacing with most common 'TA'- TA	Typical/Average
combined_data['KitchenQual'] = combined_data['KitchenQual'].fillna(combined_data['KitchenQual'].mode()[0])
#replacing with common string 
combined_data['Exterior1st'] = combined_data['Exterior1st'].fillna(combined_data['Exterior1st'].mode()[0])
combined_data['Exterior2nd'] = combined_data['Exterior2nd'].fillna(combined_data['Exterior2nd'].mode()[0])
#MSSubClass : MS SubClass (Nominal): Identifies the type of dwelling involved in the sale.
#Na most likely means no building, replacing missing values with None
combined_data['MSSubClass'] = combined_data['MSSubClass'].fillna("None")
combined_data.Functional.value_counts().sort_index()
combined_data['Functional'] = combined_data['Functional'].fillna(combined_data['Functional'].mode()[0])
```

```python
numericals = combined_data.dtypes[combined_data.dtypes != 'object'].index
# Check the skew of all numerical features
skewed_features = combined_data[numericals].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkewness in numerical features: \n")
skewness = pd.DataFrame({'skewness' :skewed_features})
skewness.head(25)
```

```python
# # Inspired by Alexandru Papiu's script : https://www.kaggle.com/apapiu/house-prices-advanced-regression-techniques/regularized-linear-models
skewness = skewness[abs(skewness) > 0.50]
print("There are {} skewed numerical features that have to be box-cox transformed".format(skewness.shape[0]))
from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15 # lam zero will give simple log transformation
for features in skewed_features:
    combined_data[features] = boxcox1p(combined_data[features], lam)
```

```python
#defining cross-validation
```

```python
n_folds = 5
def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(train_data)
    rmse= np.sqrt(-cross_val_score(model, train_data, y_train, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

```python
#LASSO Regression :
# sensitive to outliers, hence, using the sklearn's Robustscaler() method on pipeline
```

```python
lasso_model = make_pipeline(RobustScaler(), Lasso(alpha =0.0005, random_state=1))
score = rmsle_cv(lasso_model)
print("\nLasso score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))

lasso_model.fit(train_data,y_train)
# get predictions
predictions = np.exp(lasso_model.predict(test_data))
predictions

# save predictions in Kaggle-friendly format
solutions = pd.DataFrame(data = {'Id':test_ID,'SalePrice':predictions})
solutions.to_csv('lasso_model_predict.csv',index=False)
```
